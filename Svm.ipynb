1: What is a Support Vector Machine ?

 ans-

    A Support Vector Machine is a supervised machine-learning algorithm that finds the optimal hyperplane that best separates classes by maximizing the margin between them.

2: What is the difference between Hard Margin and Soft Margin SVM?

 ans-

    Difference between Hard Margin and Soft Margin SVM

  Hard Margin SVM

    No misclassifications allowed

    Assumes data is perfectly separable

    Very sensitive to noise

  Soft Margin SVM

    Allows misclassifications

    Uses penalty parameter C

    Works with noisy, overlapping data

3:What is the mathematical intuition behind SVM?

ans-

    SVM maximizes the margin, defined as the distance between the hyperplane and the nearest data points. The optimization problem:

    Minimize
      ∣w∣∣2
    
    subject to
      
      yi​(w⋅xi​+b)≥1

      Maximizing margin = minimizing ∣w∣∣.

4:What is the role of Lagrange Multipliers in SVM?

ans-

    The role of lagrange multipliers in SVM are-
  
    They convert a constrained optimization problem into a dual form.
    They help identify support vectors:
    Only points with non-zero Lagrange multipliers become support vectors.

5:What are Support Vectors?

ans-

    Support vectors are the data points closest to the decision boundary.
    They define the margin and the optimal hyperplane.

6:What is a Support Vector Classifier (SVC)?

ans-

    A Support Vector Classifier is the soft-margin version of SVM used for classification when data is not perfectly separable.

7:What is a Support Vector Regressor (SVR)?

ans-

    SVR uses SVM principles for regression.
    It tries to fit the best function within an epsilon-insensitive margin.

8:What is the Kernel Trick?

ans-

    Kernel trick transforms data into a higher-dimensional space without explicit computation.
    Allows SVM to separate non-linear data.    

9: Compare Linear Kernel, Polynomial Kernel, and RBF Kernel

ans-

    Linear: good for linearly separable data, fast
    Polynomial: captures curved boundaries, slower
    RBF: most powerful; handles complex non-linear patterns

10:What is the effect of the C parameter in SVM?

ans-

    Effect of the C parameter in SVM-

    Large C: low margin, fewer misclassifications

    Small C: large margin, more misclassifications allowed

11:What is the role of the Gamma parameter in RBF Kernel SVM?

ans-

    Role of Gamma in RBF Kernel

    High gamma: decision boundary tight around points (overfitting)

    Low gamma: smoother boundary (underfitting)

   
12: What is the Naïve Bayes classifier, and why is it called "Naïve"?

ans-

13: What is Bayes’ Theorem?

ans-

    Bayes’ Theorem is a fundamental probability formula used to calculate the probability of an event based on prior knowledge of related conditions.

        Formula

        P(A∣B)=P(B)P(B∣A)P(A)

        Where,

    P(A∣B) → Probability of event A given event B has happened (posterior probability)
    P(B∣A) → Probability of observing B if A is true (likelihood)
    P(A) → Probability of A before observing B (prior)
    P(B) → Total probability of observing B (evidence)

21: Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy

22:Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then

24:Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary:

```python
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
import numpy as np

X, y = make_blobs(n_samples=200, centers=2, random_state=42)
clf = SVC(kernel='poly', degree=3).fit(X, y)

# Plot
plt.scatter(X[:,0], X[:,1], c=y)
ax = plt.gca()
xx = np.linspace(-10, 10, 100)
yy = np.linspace(-10, 10, 100)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, levels=[0])
plt.show()
```

27:Write a Python program to train an SVM Classifier with different C values and compare the decision
boundaries visually

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=200, centers=2, random_state=42)

C_values = [0.1, 1, 10]

plt.figure(figsize=(12, 4))

for i, C in enumerate(C_values):
    clf = SVC(kernel='linear', C=C).fit(X, y)

    plt.subplot(1, 3, i+1)
    plt.title(f"C = {C}")
    plt.scatter(X[:, 0], X[:, 1], c=y)

    ax = plt.gca()
    xx = np.linspace(-10, 10, 100)
    yy = np.linspace(-10, 10, 100)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = clf.decision_function(xy).reshape(XX.shape)
    ax.contour(XX, YY, Z, levels=[0])
plt.show()
```

29: Write a Python program to apply feature scaling before training an SVM model and compare results with
unscaled data=

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

data = load_wine()
X, y = data.data, data.target

# Without scaling
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
clf = SVC().fit(X_train, y_train)
print("Without scaling:", accuracy_score(y_test, clf.predict(X_test)))

# With scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
clf2 = SVC().fit(X_train, y_train)
print("With scaling:", accuracy_score(y_test, clf2.predict(X_test)))
```

30:Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and
after Laplace Smoothing=

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = make_classification(n_samples=500, n_features=10, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model1 = GaussianNB()
model1.fit(X_train, y_train)
pred1 = model1.predict(X_test)

model2 = GaussianNB(var_smoothing=1e-6)
model2.fit(X_train, y_train)
pred2 = model2.predict(X_test)

print("Before Smoothing:", accuracy_score(y_test, pred1))
print("After Smoothing:", accuracy_score(y_test, pred2))
```

31:Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C,

38: Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the
average accuracy

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import numpy as np

X, y = load_iris(return_X_y=True)

kf = StratifiedKFold(n_splits=5)
scores = []

for train, test in kf.split(X, y):
    clf = SVC().fit(X[train], y[train])
    scores.append(accuracy_score(y[test], clf.predict(X[test])))

print("Average accuracy:", np.mean(scores))
```

40:Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
compare accuracy

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_wine(return_X_y=True)

model = SVC(kernel='linear')
selector = RFE(model, n_features_to_select=5)
X_new = selector.fit_transform(X, y)

X_train, X_test, y_train, y_test = train_test_split(X_new, y)
clf = SVC().fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, clf.predict(X_test)))
```

46:Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

X, y = make_classification(n_samples=500)
clf = SVC(probability=True).fit(X, y)

probs = clf.predict_proba(X)[:, 1]
precision, recall, _ = precision_recall_curve(y, probs)

plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()
```
